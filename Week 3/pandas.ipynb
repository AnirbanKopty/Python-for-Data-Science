# Copying Data

There are two types of copying data
1. Shallow Copy
2. Deep Copy

## 1. Shalow Copy

Here, the copied data is linked to the original data, i.e. any changes to copied data will reflect on the original data

### Syntax:
`data2 = data1`  
`data2 = data1.copy(deep=False)`

## 2. Deep Copy

Here, the copied data is a separate new database, unlinked to the previous database

### Syntax:

`data2 = data1.copy()`

> By default the `deep` parameter is set to `True`

## Two-way Table - Joint Probability

Likelihood of two events happening at the same time.

We basically convert the values into probabilities

````python
# Adding parameter `normalize=True`
pd.crosstab(columns=cars_data2.FuelType, index=cars_data2.Automatic, dropna=True, normalize=True)
````

````python
# So that you can see what it does
1104/(1104+144+15+73), 73/(1104+144+15+73)
````

i.e. What we can infer from the above data is, out of all the data available, you won't find any car likely to have been running on 'Diesel' and having 'Automatic' gear box.

## Two-way Table - Marginal Probability

The probability of the occurance of the single event

````python
# adding parameter `margins=True`
pd.crosstab(columns=cars_data2.FuelType, index=cars_data2.Automatic, dropna=True, normalize=True, margins=True)
````

This basically does the sums of column and row, so does we can see the total probability of each column or rows

## Two way Table - Conditional Probability

Given event A has happened, what's the probability of ocurring event B

Here we find the probability of different fuel type, given the type of gear box

````python
# normalize= 'index'
# So, we are making normalizarion wrt index i.e. 'Automatic' (gear type which is given)
pd.crosstab(index= cars_data2.Automatic, columns= cars_data2.FuelType, margins=True, dropna=True, normalize='index')
````

Row sum is 1, you can see, which is expected as we want to see probability of having FuelType (say Petrol for example) having GearBox (0 - Manual say)

## Correlation

To see the correlation between the points in scatter plot

Synatx is : `DataFrame.corr(self, method='pearson')`

Since the correlation function does the finding of correlation of all the values of the DataFrame objects, that's why it's a method rather than a function

The above syntax is for pairwise correlation of columns excluding NaN values
pearson correlation is the method used by default for finding correlations - used to check the strength of association between numerical variable

some Methods are

### Interpretation

- So, why are there 1(s) in the diagonal?
  - As, the correlation of the same variable with itself is 1 i.e. they are fully related
- What are those values other than 1?
  - A value above 0.5 can be considered as correlated variables (between the cell's row and column header), >0.8 mean strong correlation
- And the sign?
  - The sign represents the trend, positive means the correlated variable is 'proportionally related' to the variable, negative means 'inversely proportionally related' (The quote means we don't know the relation, we just have some idea of proportionality i.e. if one goes up what the correlated varible does)

> Note that, the matrix is symmetric, which is obvious since, the column and the row headers are same

this is a DataFrame consisting of only Missing Values contained rows

## Tackling Missing Values

No pattern to the missing values so that we can drop off missing value data (patterns like one data having multiple missing values - i.e. say more than 2-3)

If we are to remove all those at least one missing valued data, then we lose 340 observartions / 1436 obs

2 ways of filling the missing values
- Fill the missing values by `mean/median` , in case of numerical variables
- Fill the missing values by `mode` i.e. whichever class has the maximum count, in case of categorical variables

### Numerical variables

We have to be careful whether to take the Mean or the Median  
Say, if we have any extreme values in some data, that might cause the mean value to be very high/, you should thus mostly go with the median
  since, median sorts all the values in ascending/descending order, then takes the centred value


````python
# summarize the central tendency, dispersion and shape of dataset distribution excluding all the NaN values
cars_data4.describe()
````

So, we get here the total count, mean, std and the 5-number summary for all the numerical variables (with 50% being Median)

It can be noticed easily that median is an integer, whereas Mean is float, meaning the median is a value of the variable that exists

#### Missing values of Age variable

````python
# calculating the mean of the 'Age' variable
cars_data4.Age.mean()
````

````python
# To fill the missing values with the mean values for the 'Age' variable
# variable_to_fill_missing_values.fillna(<number_to_fill>, inplace=bool)
cars_data4.Age.fillna(cars_data4.Age.mean(), inplace=True)
````

````python
cars_data4.info()
````

#### Missing values of 'KM' variable

````python
# inputting the median of KM to the NaN values
cars_data4.KM.fillna(cars_data4.KM.median(), inplace=True)
````

````python
cars_data4.KM.count()
````

#### Missing values of 'HP' variable

````python
# inputting the mean value (since it is close to median)
cars_data4.HP.fillna(cars_data4.HP.mean(), inplace=True)
````

````python
cars_data4.HP.count()
````

### Categorical variables

We fill the missing values here with the most occuring categorical value

## Filling the Missing values in one shot
